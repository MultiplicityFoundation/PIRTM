# 🤖 Example QAI Integration with PIRTM

This notebook embeds **PIRTM dynamics** into a simulated **Quantum-Adaptive Intelligence (QAI)** cognitive loop, showcasing:

- Recursive memory updates
- Semantic feedback loops
- Spectral diagnostics for cognitive phase-locking

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import eigvals
from sympy import primerange
```

## 🧠 Tensor Initialization

```python
def prime_tensor(p, dim=4):
    Q, _ = np.linalg.qr(np.random.randn(dim, dim))
    return np.log(p) * Q
```

## 🔁 QAI Recursive Update Rule

```python
def evolve_qai(Xi, T, Lambda_m=0.85, feedback=None):
    if feedback is None:
        feedback = np.zeros_like(Xi)
    return Lambda_m * Xi @ T + feedback
```

## 🧬 Cognitive Feedback Operator

```python
def semantic_feedback(t, dim):
    return np.eye(dim) * np.sin(t / 5.0)
```

## 📈 Entropy & Coherence Metrics

```python
def spectral_entropy(eigs):
    mag = np.abs(eigs)**2
    mag /= np.sum(mag)
    return -np.sum(mag * np.log(mag + 1e-9))

def phase_coherence(eigs):
    return 1 - np.abs(np.sum(np.exp(1j * np.angle(eigs)))) / len(eigs)
```

## ▶️ Simulate QAI Cognition

```python
dim = 4
Xi = np.eye(dim)
T = prime_tensor(17, dim)
Lambda_m = 0.88
entropies = []
coherences = []

for t in range(40):
    Phi = semantic_feedback(t, dim)
    Xi = evolve_qai(Xi, T, Lambda_m, Phi)
    eigs = eigvals(Xi)
    entropies.append(spectral_entropy(eigs))
    coherences.append(phase_coherence(eigs))
```

## 📊 Results: Cognitive Diagnostics

```python
fig, axs = plt.subplots(2, 1, figsize=(8, 6))

axs[0].plot(entropies, label='Entropy')
axs[0].set_ylabel("Entropy")
axs[0].legend()
axs[0].grid(True)

axs[1].plot(coherences, label='Coherence', color='green')
axs[1].set_ylabel("Coherence")
axs[1].set_xlabel("Time Step")
axs[1].legend()
axs[1].grid(True)

plt.suptitle("QAI Cognitive Metrics under PIRTM Feedback")
plt.tight_layout()
plt.show()
```
