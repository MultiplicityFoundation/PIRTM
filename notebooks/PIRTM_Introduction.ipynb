```python
# 📦 Dependencies
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigvals
from sympy import primerange
from numpy.linalg import norm
```

## 🔢 1. Prime-Indexed Tensor Construction
We define a prime-indexed tensor as a matrix whose eigenvalues are structured by prime numbers.

```python
def generate_prime_tensor(n):
    primes = list(primerange(0, n*n + 1))[:n*n]
    tensor = np.array(primes).reshape((n, n))
    return tensor
```

## 🔁 2. Recursive Operator Ξ(t)
Define the recursive update rule using PIRTM feedback with control constants Λₘ and α.

```python
def recursive_update(Xi, M, alpha=0.9, beta=-0.5, p=2):
    M_t = Xi @ M
    R = M_t - M
    Xi_next = alpha * p**beta * np.outer(M_t, M_t) + np.tanh(R)
    return Xi_next, M_t
```

## 📉 3. Spectral Analysis of Prime Tensor
We analyze the eigenvalue spectrum and verify rigidity behavior predicted by the PEOH.

```python
tensor = generate_prime_tensor(10)
spectrum = eigvals(tensor)
plt.plot(np.real(spectrum), np.imag(spectrum), 'bo')
plt.title("Prime Tensor Spectrum")
plt.xlabel("Real")
plt.ylabel("Imaginary")
plt.grid(True)
plt.show()
```

## 🧪 4. Simulating Ξ(t) Evolution
Track recursive evolution to explore convergence and resonance behavior.

```python
Xi = np.eye(2)
M = np.array([1.0, 0.0])
trajectories = [Xi.copy()]
for t in range(20):
    Xi, M = recursive_update(Xi, M)
    trajectories.append(Xi.copy())
    print(f"t={t}, norm={norm(Xi)}")
```
