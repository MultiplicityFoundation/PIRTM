# 🧠 QAI_with_PIRTM.ipynb

This notebook explores the integration of **Prime-Indexed Recursive Tensor Mathematics (PIRTM)** into Quantum-Adaptive Intelligence (QAI) systems. It focuses on recursive cognitive updates, quantum feedback loops, and spectral learning trajectories.

## Notebook Structure
- Initialize QAI State Tensor
- Define Prime Evolution Operators
- Simulate Recursive Cognition Loop
- Visualize Entropy and Coherence Over Time

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import eigvals, norm
from sympy import primerange
```

## 🔧 Define Prime Evolution Tensor

```python
def prime_matrix(n):
    primes = list(primerange(2, 1000))[:n*n]
    return np.diag(primes[:n])
```

## 🔁 Recursive QAI Cognition Update
Based on:
\[ 
\Xi(t+1) = \Lambda_m \cdot \Xi(t) \cdot T_p + \Phi(t)
\]

```python
def feedback_operator(t, dim):
    phase = np.exp(1j * 2 * np.pi * np.sin(t / 10))
    return np.eye(dim) * phase

def evolve_QAI(Xi, T, Lambda_m=0.85, feedback=None):
    if feedback is None:
        feedback = np.zeros_like(Xi)
    return Lambda_m * Xi @ T + feedback
```

## ▶️ Run QAI Recursive Evolution

```python
dim = 5
Xi = np.eye(dim)
T = prime_matrix(dim)
entropies = []
coherences = []

for t in range(50):
    Phi = feedback_operator(t, dim)
    Xi = evolve_QAI(Xi, T, feedback=Phi)
    eigs = eigvals(Xi)
    probs = np.abs(eigs)**2 / np.sum(np.abs(eigs)**2)
    entropy = -np.sum(probs * np.log(probs + 1e-10))
    coherence = 1 - np.abs(np.sum(np.exp(1j * np.angle(eigs)))) / len(eigs)
    entropies.append(entropy)
    coherences.append(coherence)
```

## 📊 Cognitive Entropy and Phase Coherence

```python
fig, ax = plt.subplots(2, 1, figsize=(8, 6))

ax[0].plot(entropies, label='Spectral Entropy')
ax[0].set_ylabel("Entropy")
ax[0].legend()
ax[0].grid(True)

ax[1].plot(coherences, label='Phase Coherence')
ax[1].set_ylabel("Coherence")
ax[1].set_xlabel("Time Step")
ax[1].legend()
ax[1].grid(True)

plt.tight_layout()
plt.show()
```
