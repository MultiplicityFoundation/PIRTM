# 🔄 RecursiveTensor_EigenSimulation.ipynb

This notebook simulates recursive tensor evolution using **PIRTM operators** and visualizes the **eigenvalue trajectories** under dynamic feedback.

## Contents
- Initialize Recursive Tensor System
- Apply Prime-Indexed Evolutions
- Visualize Eigenvalue Drift
- Analyze Convergence and Phase Stability

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import primerange
from numpy.linalg import eigvals, norm
```

## 🔢 Prime Tensor Generator

```python
def prime_tensor(n):
    primes = list(primerange(2, 1000))[:n**2]
    return np.array(primes).reshape((n, n)) * np.eye(n)
```

## 🧮 Recursive Evolution Function

```python
def evolve_tensor(Xi, T, Lambda_m=0.88, feedback=None):
    if feedback is None:
        feedback = np.zeros_like(Xi)
    return Lambda_m * Xi @ T + feedback
```

## 🔁 Run Evolution and Capture Spectra

```python
dim = 4
Xi = np.eye(dim)
T = prime_tensor(dim)
spectra = []

for t in range(40):
    Xi = evolve_tensor(Xi, T, Lambda_m=0.88)
    eigs = eigvals(Xi)
    spectra.append(eigs)
```

## 🎯 Visualize Eigenvalue Trajectories

```python
for step, eigs in enumerate(spectra):
    plt.scatter(eigs.real, eigs.imag, alpha=0.5, label=f"t={step}" if step % 10 == 0 else "")

plt.title("Recursive Eigenvalue Evolution")
plt.xlabel("Re(λ)")
plt.ylabel("Im(λ)")
plt.grid(True)
plt.legend()
plt.axis('equal')
plt.show()
```

## 📊 Stability Diagnostics
Estimate convergence by tracking the Frobenius norm of Ξ(t).

```python
norms = [norm(Xi)]
Xi = np.eye(dim)

for t in range(40):
    Xi = evolve_tensor(Xi, T)
    norms.append(norm(Xi))

plt.plot(norms)
plt.title("Ξ(t) Norm Over Time")
plt.xlabel("Time Step")
plt.ylabel("||Ξ(t)||")
plt.grid(True)
plt.show()
```
